```python
#count the number of numerical and none-numerical columns
from pandas.api.types import is_numeric_dtype

n=0
for c in data.columns:
    if is_numeric_dtype(data[c]):
        n=n+1
        
print(f"numerical columns: {n}, and none-numerical columns {len(data.columns) - n}: ")                    
```

```python
#clean abnormal values
data['Volume'].min()
```

```python
data['Volume'].max()
```

```python
data['Volume'].sort_values(ascending=True).head(20)
```

```python
data=data[data['Volume'] > 100000]
```

```python
data['Volume'].sort_values(ascending=True).head(20)
```

```python
data
```

```python
# Remove all columns that have a constant/single value
constant_columns=[]
for c in data.columns:
    n = len(data[c].unique())
    if n == 1:
        constant_columns.append(c)
        

print(f" Constant columns are: {constant_columns}")        
```

```python
data['Source'].unique()
```

```python
data.drop(columns=constant_columns, axis=1, inplace=True)
```

```python
data
```

```python
# Adding Year, Month, Dat columns to the data
data.info()
```

```python
dateObject = pd.DatetimeIndex(data['Date'])
dateObject
```

```python
dateObject.day
```

```python
data['Day'] = dateObject.day
data['Month'] = dateObject.month
data['Year'] = dateObject.year

data
```

```python
data[['Date','Day','Month','Year']]
```

```python
# Transform the values of Volume entries to represent thousands
data['Volume'] = data['Volume'] / 1000
```

```python
data
```

```python
#Data visualization for analysis
data[['Year', 'Volume']].groupby('Year').sum().astype(int)
```

```python
data[['Year','Volume']].groupby('Year').sum().astype(int).plot(figsize=(8,6), y='Volume', fontsize=12, title="Stock's Volume Over Years")
plt.ylabel("Stock's Volume")
```
